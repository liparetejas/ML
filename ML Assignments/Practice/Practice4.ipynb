```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
curr_x = 25
max_iterations = 10000
g = lambda x : 2*(x+3)
iters = 1
rate = 0.01
precision = 0.0000001
prev_step_size = 1
```

```python
y = [(x+3)**2 for x in range(-100, 100)]
x = [x for x in range(-100, 100)]
```

```python
sns.lineplot(x, y)
```

```python
progression_x = [curr_x]
while prev_step_size > precision and iters < max_iterations:
  prev_x = curr_x
  curr_x -= rate * g(curr_x)
  prev_step_size = abs(curr_x-prev_x)
  iters += 1
  progression_x.append(curr_x)
  print('Iteration:', iters, '\nValue of X:', curr_x)
```

```python
progression_y = [(x+3)**2 for x in progression_x]
sns.scatterplot(progression_x, progression_y)
sns.lineplot(x, y)
```

```python
print('Local minima:', curr_x, '\nIteration:', iters)
```
